### 3. Import packages

```python
from pathlib import Path
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt 
import sklearn
```

### 4. Directory setup

```python
BASE_DIR = Path.home() / "Projects" / "BitSize_ML"
PROJECT_NAME = "end_to_end_ml"
PROJECT_DIR = BASE_DIR / PROJECT_NAME
```

```python
SUB_DIR = {
    "images":"images",
    "data":"data",
    "notebooks":"notebooks",
    "models":"models"
}
```

```python
PATHS = {name:(PROJECT_DIR/rel).resolve() for name, rel in SUB_DIR.items()}
PATHS
```

```python
for p in PATHS.values():
    p.mkdir(parents=True, exist_ok=True)
```

### 5. Custom figure saving

```python
def save_fig(name:str, ext:str="png", dpi:int=300, tight:bool=True):
    path = PATHS["images"]/f"{name}.{ext}"
    if tight:
        plt.tight_layout()
    plt.savefig(path, format=ext, dpi=dpi)
    print("[Saved]",path)
```

### 6. Quick sanity check

```python
x = np.linspace(0, 2*np.pi, 200)
y = np.sin(x)

plt.figure()
plt.plot(x,y)
plt.title("Checking")
plt.xlabel("x")
plt.ylabel("sin(x)")
save_fig("ENV checking")
plt.show()
plt.close()
```
